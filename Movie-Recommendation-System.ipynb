```python
import numpy as np 
import pandas as pd 
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv') 
```

```python
movies = movies.merge(credits,on='title')
movies.head()
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies.dropna(inplace=True)
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
#movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.sample(5)
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
#new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
    
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
new[new['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
def recommend(movie):
    movie = movie.lower()
    titles = movies['title'].str.lower()

    # Handle movie not found
    if movie not in titles.values:
        return [" Movie not found! Please check the spelling (e.g., 'Avatar')"]

    try:
        idx = titles[titles == movie].index[0]
        distances = list(enumerate(similarity[idx]))
        sorted_movies = sorted(distances, key=lambda x: x[1], reverse=True)[1:6]
        return [movies.iloc[i[0]].title for i in sorted_movies]
    except Exception as e:
        return [f"⚠️ Error: {str(e)}"]
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output

# Input field
movie_input = widgets.Text(
    value='Avatar',
    placeholder='Type a movie name',
    description='Movie:',
    disabled=False
)

output = widgets.Output()

def on_submit(change):
    with output:
        clear_output()
        movie = change['new']
        recommendations = recommend(movie)
        print(f"🎬 Top 5 recommendations for '{movie.title()}':")
        for rec in recommendations:
            print("•", rec)

movie_input.observe(on_submit, names='value')

display(movie_input, output)
```
